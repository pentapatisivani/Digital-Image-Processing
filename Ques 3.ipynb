```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
plt.rcParams["figure.figsize"] = (20,20)
```

```python
def convolve(img,kernel):
    k_size=kernel.shape
    i_size=img.shape
    img1=np.zeros((i_size[0],i_size[1]))
    image_padded = np.zeros((i_size[0] + 2, i_size[1] + 2))   
    image_padded[1:-1, 1:-1] = img
    for i in range(i_size[0]):
        for j in range(i_size[1]):
            temp=np.multiply(kernel,image_padded[i:i+3,j:j+3])
            img1[i,j]=temp.sum()
   
    img1=np.clip(img1,0,255)        
    return img1
    
    
```

```python
box_img=cv2.imread('box.png',0)
k = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
k=np.flip(k)
op=convolve(box_img,k)
plt.subplot(1,2,1)
plt.imshow(box_img,cmap='gray')
plt.title("Input Image")
plt.subplot(1,2,2)
plt.imshow(op,cmap='gray')
plt.title("Output Image")
plt.show()
```

## Part 2:

```python
blur_img=cv2.imread('blur.jpg',0)
k1=k.T
op1=convolve(blur_img,k)
op2=convolve(blur_img,k1)
plt.subplot(3,1,1)
plt.imshow(blur_img,cmap='gray')
plt.title("Input Image")
plt.subplot(3,1,2)
plt.imshow(op1,cmap='gray')
plt.title("Output Image")
plt.subplot(3,1,3)
plt.imshow(op2,cmap='gray')
plt.title("Ouput with its transpose")
plt.show()
```
