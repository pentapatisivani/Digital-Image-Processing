## Image Restoration

### From  the given input noise image and output denoised, image, it can be observed that the filter used to denoise the given image will be a median filter as it not just removes noise but also preserves edges(since it can be observed that the output image had it's edges preserved). Therefore the appropriate edge preserving noise removing median filter is used


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy as np

```

```python
plt.rcParams["figure.figsize"] = (20,20)
```

```python
def medianfilter(img,ker):
    size=img.shape
    new_img=img.copy()
    for i in range(1,size[0]-1):
        for j in range(1,size[1]-1):
            temp=img[i-1:i+2,j-1:j+2,:]
            new_img[i][j][0]=np.median(temp[:,:,0])
            new_img[i][j][1]=np.median(temp[:,:,1])
            new_img[i][j][2]=np.median(temp[:,:,2])
    
    return new_img
```

```python
i=cv2.imread('Degraded.jpg',cv2.IMREAD_COLOR)
i = cv2.cvtColor(i, cv2.COLOR_BGR2RGB)
a=medianfilter(i,3)
plt.subplot(1,2,1)
plt.imshow(i.astype('uint8'))
plt.title("Input noisy image")
plt.subplot(1,2,2)
plt.imshow(a.astype('uint8'))
plt.title("Median filtered output image with kernel size:3")
plt.show()
```
